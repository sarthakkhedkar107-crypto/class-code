```python
total=int(input("Total bill"))

if total>100:
    discount=total*0.2
    finalbill=total-discount
    print(f'total was {total} we are gving dicount {discount} final price{finalbill}')
elif total>50:
    discount=total*0.1
    finalbill=total-discount
    print(f'total was {total} we are gving dicount {discount} final price{finalbill}')
else:
   print(f'total was {total} we dont have discount')
```
