 # while

```python
age = 15
while age < 18:
    print('Student',age)
    age = age + 1
```

```python
time=0
while time <=10:
    print(time)
    time=time+1
```

```python
button='hellow'
while button!='exit':
    print(button)
    button=input("Enter the value")
```

 # BREAK

```python
for i in range(1,10):
    if i==4:
        break
    else:
        print(i)
     
       
    
```

```python
for i in 'hellow':
    if i=='l':
        break
    else:
        print(i)
```

# continue


```python
for i in range(1,10):
    if i==4:
        continue
    else:
        print(i)
     
```

```python
for i in 'hellow':
    if i=='l':
        continue
    else:
        print(i)
```
